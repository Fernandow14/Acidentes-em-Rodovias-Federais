```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path("..")
DATA_DIR = BASE_DIR / "data" / "raw"

df_2023 = pd.read_csv(DATA_DIR / "acidentes2023_todas_causas.csv", sep=";", encoding="latin1")
df_2024 = pd.read_csv(DATA_DIR / "acidentes2024_todas_causas.csv", sep=";", encoding="latin1")
df_2025 = pd.read_csv(DATA_DIR / "acidentes2025_todas_causas.csv", sep=";", encoding="latin1")
```

```python
import pandas as pd

df_2023 = pd.read_csv(
    "../data/raw/acidentes2023_todas_causas.csv",
    sep=";",
    encoding="latin1"
)

df_2024 = pd.read_csv(
    "../data/raw/acidentes2024_todas_causas.csv",
    sep=";",
    encoding="latin1"
)

df_2025 = pd.read_csv(
    "../data/raw/acidentes2025_todas_causas.csv",
    sep=";",
    encoding="latin1"
)
```

```python
#Data

for df in [df_2023, df_2024, df_2025]:
    df["data_inversa"] = pd.to_datetime(
        df["data_inversa"],
        errors="coerce"
    )
```

```python
#KM
for df in [df_2023, df_2024, df_2025]:
    df["km"] = pd.to_numeric(
        df["km"],
        errors="coerce"
    )
```

```python
#Longitude

for df in [df_2023, df_2024, df_2025]:
    if "longitude" in df.columns:
        df["longitude"] = pd.to_numeric(
            df["longitude"],
            errors="coerce"
        )
```

```python
#latitude

for df in [df_2023, df_2025]:
    if "latitude" in df.columns:
        df["latitude"] = pd.to_numeric(
            df["latitude"],
            errors="coerce"
        )
```

```python
#pesid

for df in [df_2023, df_2024, df_2025]:
    if "pesid" in df.columns:
        df["pesid"] = pd.to_numeric(
            df["pesid"],
            errors="coerce"
        ).astype("Int64")
```

```python
#id_veiculos

for df in [df_2023, df_2024, df_2025]:
    if "id_veiculo" in df.columns:
        df["id_veiculo"] = pd.to_numeric(
            df["id_veiculo"],
            errors="coerce"
        ).astype("Int64")
```

```python
#ano_fabricacao_veiculo

for df in [df_2023, df_2024, df_2025]:
    if "ano_fabricacao_veiculo" in df.columns:
        df["ano_fabricacao_veiculo"] = pd.to_numeric(
            df["ano_fabricacao_veiculo"],
            errors="coerce"
        ).astype("Int64")
```

```python
#Alterando colunas de nulos para não informado


colunas_categoricas = [ #definindo as colunas categóricas para preencher os nulos com "Não informado"
    "classificacao_acidente",
    "regional",
    "delegacia",
    "uop",
    "tipo_veiculo",
    "tipo_envolvido",
    "condicao_metereologica"
]

for df in [df_2023, df_2024, df_2025]: # aplicando o tratamento nos 3 anos 
    for col in colunas_categoricas: # percorrendo as variáveis categoria
        if col in df.columns: #verificando possiveis erro nos anos caso exista dados diferentes
            df[col] = df[col].fillna("Não informado")
```

```python
# data_inversa / uf

for df in [df_2023, df_2024, df_2025]:
    df.dropna(subset=["data_inversa", "uf"], inplace=True)
```

```python
#conferindo os dataframes tratados

df_2023.shape, df_2024.shape, df_2025.shape
```

```python
#criando tabela de ano

df_2023["ano"] = 2023
df_2024["ano"] = 2024
df_2025["ano"] = 2025

```

```python
#conferindo tabela de ano

print("2023:", df_2023["ano"].unique())
print("2024:", df_2024["ano"].unique())
print("2025:", df_2025["ano"].unique())
```

```python
#unindo os anos 

df_final = pd.concat(
    [df_2023, df_2024, df_2025],
    ignore_index=True
)
```

```python
#mudando o nome da coluna pesid para id_pessoa

df_final.rename(columns={"pesid": "id_pessoa"}, inplace=True)
```

```python
#mudando o nome das colunas feridos para acidentes

df_final.rename(columns={"feridos_leves": "Acidentes_leves"}, inplace=True)
df_final.rename(columns={"feridos_graves": "Acidentes_graves"}, inplace=True)
```

```python
#mudando o nome da coluna uop para código
df_final.rename(columns={"uop": "Código"}, inplace=True)
```

```python
#analisando as colunas tratadas com dados alterados, verifiquei que a coluna id está como float e não inteiro, então farei a alteração para Int64 novamente.

df_final["id"] = (
    pd.to_numeric(df_final["id"], errors="coerce")
    .astype("Int64")
)
```
